```python
import pandas as pd
import os
```

### data_file will not be kept in repo. Too large for github upload.

```python
# Assign data file path
data_file = '../../assets/data/raw_samples/query_181021%2F000000000000.csv'
sample_file = '../../assets/data/raw_samples/query_181021_sample.csv'
chunk_size = 10 ** 6

df_head = pd.read_csv(data_file).sample(n=20)

df_head.head(5)
```

```python
# Chunk and skip half the data
df_chunked = pd.read_csv(data_file, skiprows=lambda x: x % 2, chunksize=chunk_size)

# Write the kept half to a sample csv file
for chunk in df_chunked:
    if os.path.isfile(sample_file):  # check if file already exists
        chunk.to_csv(sample_file, mode='a', header=False)
    else:
        chunk.to_csv(sample_file)
```
